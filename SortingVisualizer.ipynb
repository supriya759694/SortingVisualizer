```python
def visualize_sort(array, sort_function, title):
    fig, ax = plt.subplots()
    bar_rects = ax.bar(range(len(array)), array, align="edge")
    ax.set_title(title)
    ax.set_xlim(0, len(array))
    ax.set_ylim(0, max(array) + 1)
    plt.pause(0.5)
    sort_function(array, bar_rects, ax)
    plt.show()
```

```python
def update_bar_rects(bar_rects, array):
    for rect, value in zip(bar_rects, array):
        rect.set_height(value)
    plt.pause(0.01)
```

```python
# Bubble Sort Visualization
def bubble_sort(arr, bar_rects, ax):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                update_bar_rects(bar_rects, arr)
    update_bar_rects(bar_rects, arr)
```

```python
# Selection Sort Visualization
def selection_sort(arr, bar_rects, ax):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        update_bar_rects(bar_rects, arr)
    update_bar_rects(bar_rects, arr)
```

```python
# Insertion Sort Visualization
def insertion_sort(arr, bar_rects, ax):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
        update_bar_rects(bar_rects, arr)
    update_bar_rects(bar_rects, arr)
```

```python
# Merge Sort Visualization
def merge_sort(arr, bar_rects, ax):
    def merge(l, r):
        result = []
        i = j = 0
        while i < len(l) and j < len(r):
            if l[i] < r[j]:
                result.append(l[i])
                i += 1
            else:
                result.append(r[j])
                j += 1
        result.extend(l[i:])
        result.extend(r[j:])
        return result

    def sort_and_merge(arr):
        if len(arr) <= 1:
            return arr
        mid = len(arr) // 2
        left = sort_and_merge(arr[:mid])
        right = sort_and_merge(arr[mid:])
        merged = merge(left, right)
        for i in range(len(arr)):
            arr[i] = merged[i]
            update_bar_rects(bar_rects, arr)
        return arr

    sort_and_merge(arr)
    update_bar_rects(bar_rects, arr)

```

```python
# Quick Sort Visualization
def quick_sort(arr, bar_rects, ax):
    def partition(low, high):
        pivot = arr[high]
        i = low - 1
        for j in range(low, high):
            if arr[j] < pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
                update_bar_rects(bar_rects, arr)
        arr[i + 1], arr[high] = arr[high], arr[i + 1]
        update_bar_rects(bar_rects, arr)
        return i + 1

    def sort(low, high):
        if low < high:
            pi = partition(low, high)
            sort(low, pi - 1)
            sort(pi + 1, high)

    sort(0, len(arr) - 1)
    update_bar_rects(bar_rects, arr)
```

```python
# Array to be sorted
array = [100, 20, 201, 120, 95, 55, 12, 80, 82, 110]
```

```python
# Generate a random array
array_length = 20
array = list(range(1, array_length + 1))
random.shuffle(array)
```

```python
#Visualize each sorting algorithm
for algorithm, title in [
    (bubble_sort, "Bubble Sort"),
    (selection_sort, "Selection Sort"),
    (insertion_sort, "Insertion Sort"),
    (merge_sort, "Merge Sort"),
    (quick_sort, "Quick Sort")
]:
    visualize_sort(array.copy(), algorithm, title)
```
